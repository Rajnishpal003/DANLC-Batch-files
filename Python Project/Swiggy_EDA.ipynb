```python
#importing all libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read csv file
df=pd.read_csv("Swiggy_dataset.csv")
```

```python
#statistical summary of data
df.describe()
```

```python
#find missing value 
df.isnull().sum()
```

```python
#dropping city column
df.drop(columns='City',inplace=True)
```

```python
df.info()
```

```python
#dropping uuid columns with missing values
df.dropna(subset=['uuid','address','locality'],inplace=True)
```

```python
df['unserviceable'].fillna('False')          
df['veg'].fillna('False') 
```

```python
df.info()
```

```python
df.dropna(subset=['unserviceable','veg'],inplace=True)
```

```python
df.isnull().sum()
```

```python
df['avgRating'] = df['avgRating'].replace('--', np.nan).astype(float)
```

```python
#checking shape of final cleaned data
df.shape
```

```python
df.to_csv('rajnishpalsingh@Mac DANLC-Batch-files/Swiggy_cleaned_dataset.csv',index=False)
```

```python
print("Cleaned file saved at path")
```

```python
#plotting graphs for visulaization
#Distribution of average rating
plt.figure(figsize=(8,5))
sns.histplot(df['avgRating'],bins=10,color='blue')
plt.title("Distribution of average rating")
plt.xlabel('Average Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
#Average rating by city analysis
avgrating_by_city=df.groupby('city')['avgRating'].mean().sort_values(ascending=True)
avgrating_by_city.plot(kind='bar',color='green')
plt.title("Average rating by city")
plt.xlabel("City")
plt.ylabel("Average rating")
```

```python
#Top 5 Popular Cuisines
cuisines_count=df['cuisines'].value_counts().head()
plt.pie(cuisines_count,labels=cuisines_count.index,autopct='%1.1f%%')
plt.title("Top 5 Popular Cuisines")
plt.show()
```

```python
#Count of Veg and Non-veg Restaurants Analysis
veg_count=df['veg'].value_counts()
veg_count.plot(kind='bar',color=['orange','green'])
plt.title("Count of Veg and Non-veg Restaurants")
plt.xlabel("Veg and Non-veg")
plt.ylabel("Count")
plt.show()
```

```python
#Delivery Time by City Analysis
plt.figure(figsize=(10,8))
sns.boxplot(x='city',y='deliveryTime',data=df)
plt.title("Delivery Time by City")
plt.xlabel("City")
plt.ylabel("Delivery Time")
plt.grid(True)
plt.show()
```

```python
df['costForTwoStrings'] = pd.to_numeric(df['costForTwoStrings'], errors='coerce')
```

```python
#Total Rating Vs Avg. Rating
plt.figure(figsize=(10,8))
plt.scatter(df['totalRatingsString'], df['avgRating'], alpha=0.6, color='green')
plt.title("Total Rating Vs Avg. Rating")
plt.xlabel("Total Rating")
plt.ylabel("Average Rating")
plt.show()
```

```python
# Delivery Time Range (Min vs. Max) Analysis
plt.figure(figsize=(14, 6))
df[['minDeliveryTime', 'maxDeliveryTime']].plot(kind='box', patch_artist=True, figsize=(10, 6))
plt.title("Delivery Time Range (Min vs. Max)", fontsize=14)
plt.ylabel("Delivery Time (minutes)")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
